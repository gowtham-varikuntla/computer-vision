**VGG16**

```python
import keras
import tensorflow as tf
from keras.models import Model
from keras.applications.vgg16 import VGG16
import numpy as np
from keras.layers import Dense, Dropout, Flatten
from keras.callbacks import EarlyStopping
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.cifar10.load_data()
```

```python
x_train_resize=[];
x_test_resize=[];
input_shape=(48,48,3)
for i in x_train:
  x_train_resize.append(np.resize(i,input_shape)/255)
for j in x_test:
  x_test_resize.append(np.resize(j,input_shape)/255)
```

```python
x_train_resize=np.array(x_train_resize)
x_test_resize=np.array(x_test_resize)
base_model=VGG16(classes=10,include_top=False,input_shape=input_shape)
base_model.summary()
```

```python
for layer in base_model.layers:
    if layer.name == "block5_conv1":
        layer.trainable = True
        print("Layer " + layer.name + " not frozen")
    elif layer.name == "block5_conv2":
        layer.trainable = True
        print("Layer " + layer.name + " not frozen")
    elif layer.name == "block5_conv3":
        layer.trainable = True
        print("Layer " + layer.name + " not frozen")
    elif layer.name == "block5_conv4":
        layer.trainable = True
        print("Layer " + layer.name + " not frozen")
    elif layer.name == "block5_pool":
        layer.trainable = True
        print("Layer " + layer.name + " not frozen")
    else:
        layer.trainable = False
        print("Layer " + layer.name + " frozen")
```

```python
last=base_model.layers[-1].output
x=Dropout(0.3)(last)
x=Flatten()(x)
x=Dropout(0.3)(x)
#x=Dense(1024,activation="relu",name='fc1')(x)
#x=Dropout(0.3)(x)
x=Dense(10,activation="softmax",name='classification')(x)
model=Model(base_model.input,x)
model.summary()
```

```python
optimizer=keras.optimizers.Adam(learning_rate=0.05)
model.compile(optimizer=optimizer,loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
es=EarlyStopping(monitor='loss',mode='min',verbose=1,patience=20)
hist=model.fit(x_train_resize,y_train,epochs=150,callbacks=[es],validation_split=0.2,batch_size=32)
```

```python
y_pred=model.predict(x_test_resize)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history['accuracy'],label='train')
plt.plot(hist.history['val_accuracy'],label='validation')
plt.legend()
```
